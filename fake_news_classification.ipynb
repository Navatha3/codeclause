```python
data=pnd.concat([FAKE,TRUE],axis=0)
```

```python
data=data.reset_index(drop=True)
```

###TEXT PREPROCESSING

```python
data=data.drop(['subject','date','title'],axis=1)
```

```python
print(data.columns)
```

**TOKENIZATION**

```python
from nltk.tokenize import word_tokenize
```

```python
data['text']=data['text'].apply(word_tokenize)
```

```python
data.head()
```

**STEMMING**

```python
from nltk.stem.snowball import SnowballStemmer
English=SnowballStemmer("english")
```

```python
def stem_word(text):
  return [English.stem(word) for word in text]
```

```python
data['text']=data['text'].apply(stem_word)
```

```python
data.head()
```

**STOPWORD REMOVAL**

```python
#from nltk.corpus import stopwords
#nltk.download('stopwords')
#print(stopwords.words('english'))
```

```python
def stop_word(t):
  dt=[word for word in t if len(word)>3]
  return dt
```

```python
data['text']=data['text'].apply(stop_word)
```

```python
data.head()
```

```python
data['text']=data['text'].apply(''.join)
```

```python
data
```

**SPLITTING**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(data['text'],data['genuineness'],test_size=0.25)
```

```python
X_train.head()
```

```python
y_train.head()
```

**VECTORIZATION**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_df=0.7)
```

```python
tfidf_train=tfidf.fit_transform(X_train)
tfidf_test=tfidf.transform(X_test)
```

```python
print(tfidf_train)
```

```python
print(tfidf_test)
```

**MODEL1-LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
Model1=LogisticRegression(max_iter=900)
Model1.fit(tfidf_train,y_train)
Pred1=Model1.predict(tfidf_test)
AS_1=accuracy_score(y_test,Pred1)
```

```python
AS_1
```

**PASSIVEAGRESSIVECLASSIFIER**

```python
from sklearn.linear_model import PassiveAggressiveClassifier
Model_2=PassiveAggressiveClassifier(max_iter=100)
Model_2.fit(tfidf_train,y_train)
```

```python
Pred_2=Model_2.predict(tfidf_test)
AS_2=accuracy_score(y_test,Pred_2)
print('the accuracy score is',AS_2*100)
```
